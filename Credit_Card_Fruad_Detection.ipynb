```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

X2 = df.drop(['Class'], axis=1)
y2 = df['Class']
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=123)

dt = DecisionTreeClassifier(random_state=123)
model3 = dt.fit(X2_train, y2_train)
prediction3 = model3.predict(X2_test)
```

```python
print("Classification Report:")
print(classification_report(y2_test, prediction3))

accuracy = accuracy_score(y2_test, prediction3)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

smote = SMOTE(random_state=123)
X2_balanced, y2_balanced = smote.fit_resample(X2, y2)
X2_train_balanced, X2_test_balanced, y2_train_balanced, y2_test_balanced = train_test_split(
    X2_balanced, y2_balanced, test_size=0.3, random_state=123
)
scaler = StandardScaler()
X2_train_balanced = scaler.fit_transform(X2_train_balanced)
X2_test_balanced = scaler.transform(X2_test_balanced)

model3_balanced = dt.fit(X2_train_balanced, y2_train_balanced)
prediction3_balanced = model3_balanced.predict(X2_test_balanced)

print("Classification Report (Balanced Data):")
print(classification_report(y2_test_balanced, prediction3_balanced))
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=123)
X2_under, y2_under = undersampler.fit_resample(X2, y2)
X2_train_under, X2_test_under, y2_train_under, y2_test_under = train_test_split(
    X2_under, y2_under, test_size=0.3, random_state=123
)
model3_under = dt.fit(X2_train_under, y2_train_under)
prediction3_under = model3_under.predict(X2_test_under)

print("Classification Report (Undersampled Data):")
print(classification_report(y2_test_under, prediction3_under))
```

```python
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

def evaluate_model(y_true, y_pred, model_name):
    report = classification_report(y_true, y_pred, output_dict=True)
    accuracy = accuracy_score(y_true, y_pred)
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': report['weighted avg']['precision'],
        'Recall': report['weighted avg']['recall'],
        'F1-Score': report['weighted avg']['f1-score']
    }
results = []
results.append(evaluate_model(y2_test, prediction3, "Original Data"))
results.append(evaluate_model(y2_test_balanced, prediction3_balanced, "Oversampled Data (SMOTE)"))
results.append(evaluate_model(y2_test_under, prediction3_under, "Undersampled Data"))
results_df = pd.DataFrame(results)
print(results_df)

import matplotlib.pyplot as plt
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = range(len(metrics))

plt.figure(figsize=(10, 6))
for i, row in results_df.iterrows():
    plt.plot(x, row[metrics], label=row['Model'], marker='o')

plt.xticks(x, metrics)
plt.ylabel("Score")
plt.title("Model Comparison Across Metrics")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
X1 = df.drop(['Class'], axis=1)
y1 = df['Class']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=123)
```

```python
lr = LogisticRegression(max_iter=1000)

model2 = lr.fit(X1_train, y1_train)

prediction2 = model2.predict(X1_test)
```

```python
accuracy = accuracy_score(y1_test, prediction2)
print(f'Accuracy: {accuracy:.4f}')
```

```python
cm = confusion_matrix(y1_test, prediction2)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Fraud', 'Fraud'], yticklabels=['Non-Fraud', 'Fraud'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
precision = precision_score(y1_test, prediction2)
recall = recall_score(y1_test, prediction2)
f1 = f1_score(y1_test, prediction2)

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')
```

```python
smote = SMOTE(random_state=123)
X1_train_smote, y1_train_smote = smote.fit_resample(X1_train, y1_train)

model2_smote = lr.fit(X1_train_smote, y1_train_smote)

prediction2_smote = model2_smote.predict(X1_test)

accuracy_smote = accuracy_score(y1_test, prediction2_smote)
precision_smote = precision_score(y1_test, prediction2_smote)
recall_smote = recall_score(y1_test, prediction2_smote)
f1_smote = f1_score(y1_test, prediction2_smote)

print(f'Accuracy (SMOTE): {accuracy_smote:.4f}')
print(f'Precision (SMOTE): {precision_smote:.4f}')
print(f'Recall (SMOTE): {recall_smote:.4f}')
print(f'F1-Score (SMOTE): {f1_smote:.4f}')
```
