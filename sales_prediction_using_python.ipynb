By leveraging these predictions, businesses can optimize their advertising strategies and maximize sales potential.
```python
import matplotlib.pyplot as plt

features = ['TV', 'Radio', 'Newspaper']
target = 'Sales'

for feature in features:
    plt.figure(figsize=(6, 4))
    plt.scatter(df[feature], df[target], alpha=0.7, edgecolors='k')
    plt.title(f'{feature} vs {target}', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel(target, fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['TV'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of TV Advertising Spend', fontsize=14)
plt.xlabel('TV Advertising Spend', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['Radio'], bins=10, kde=False, color='green')
plt.title('Histogram of Radio Advertising Spend', fontsize=14)
plt.xlabel('Radio', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['Newspaper'], bins=10, kde=False, color='purple')
plt.title('Histogram of Newspaper Advertising Spend', fontsize=14)
plt.xlabel('Newspaper', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix Heatmap', fontsize=14)

for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['TV']], df[['Sales']], test_size = 0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
res= model.predict(X_test)
print(res)
```
